# Load bulk data

```python
files=["4607","8121","9692","13904","26378","26440","28750","57065","95233"]
rootpath="./dataBulk/"
norm=".txt"
sample=pd.read_csv(rootpath+"exp.gene.lncRNA.GSE"+files[0]+norm , sep="\t").T
print("sample shape",sample.shape)
label=pd.read_csv(rootpath+"label_GSE"+files[0]+'.txt' , sep="\t",header=None)
print("label shape",label.shape)
mRNA_sample=pd.read_csv(rootpath+"exp.gene.mRNA.GSE"+files[0]+norm , sep="\t").T
print("mRNA sample shape",mRNA_sample.shape)

print(mRNA_sample)
print(sample)
print(label)
for i in range(1,len(files)):
  temp_sample=pd.read_csv(rootpath+"exp.gene.lncRNA.GSE"+files[i]+norm , sep="\t").T
  temp_mRNA_sample=pd.read_csv(rootpath+"exp.gene.mRNA.GSE"+files[i]+norm , sep="\t").T
  temp_label=pd.read_csv(rootpath+"label_GSE"+files[i]+'.txt' , sep="\t",header=None)

  sample = sample.append(temp_sample)
  mRNA_sample = mRNA_sample.append(temp_mRNA_sample)
  label = label.append(temp_label)

print("sample shape",sample.shape)
print("mRNA sample shape",mRNA_sample.shape)
print("label shape",label.shape)
```

```python
sample_head=sample.index.values
mRNA_sample_head=mRNA_sample.index.values
#check if sample & mRNA_sample share same sample in same order
missing=[]
print("following sample from sample_head do not exist in mRNA_sample")
for i in range(len(sample_head)):
  if not(sample_head[i] in mRNA_sample_head[i]):
      print(sample_head[i],mRNA_sample_head[i],i)
      missing.append(i)

#double verify is it really not exist
print("double verification")
for i in range(len(missing)):
  print(mRNA_sample_head[missing[i]])

print(label[0].values.shape)
```

```python
print("remove mismatch label or sample")
for sam in sample_head:
  for each in label[0].values:
    if each in sam:
      break
  else:
    sample=sample.drop(sam)
    mRNA_sample=mRNA_sample.drop(sam)
```

```python
print("sample shape:",sample.shape)
print("mRNA_sample shape:",mRNA_sample.shape)
print("label shape:",label.shape)
```

```python
for i in range(len(sample)):
  if not(label[0].values[i] in sample.index.values[i]):
    print("order mismatch")
    exit()
```

```python
print(len(mRNA_sample.columns))
for i in range(len(mRNA_sample.columns)):
  if mRNA_sample.columns[i] in sample.columns:
    print("same column",i,mRNA_sample.columns[i])
    continue
  sample.insert(0,mRNA_sample.columns[i], mRNA_sample.iloc[:,i].to_numpy())
```

```python
RNA_used_in_previous_datase=pd.read_csv(rootpath+"rna_names.csv").T
print(RNA_used_in_previous_datase)
count=0
common_rna=[]
common_indicator=[]
for i in RNA_used_in_previous_datase:
  if RNA_used_in_previous_datase[i][0] in sample.columns:
    count+=1
    common_rna.append(RNA_used_in_previous_datase[i][0])
    common_indicator.append(True)
  else:
    common_indicator.append(False)
    #print(str(RNA_used_in_previous_datase[i][0]))

print(RNA_used_in_previous_datase.shape)
print(count)
# print(common_rna)      
df = pd.DataFrame(common_rna) 
# print(df)
# df.to_csv(rootpath+'common_rna.csv') 

# print(common_indicator)      
df = pd.DataFrame(common_indicator) 
# print(df)
# df.to_csv(rootpath+'common_rna_indicator.csv') 
common_sample=pd.DataFrame()
# print(pd.DataFrame(sample.T.iloc[0,:]).T)
print(common_sample.append(pd.DataFrame(sample.T.loc['ZZZ3',:]).T))
print(sample.shape)
# print(sample.T)
for i in common_rna:
  common_sample = common_sample.append(pd.DataFrame(sample.T.loc[i,:]).T)
common_sample=common_sample.T
# print(common_sample)
label=label[2].to_numpy()
#sample=sample.to_numpy()
sample=common_sample.to_numpy()
print("sample shape after to_numpy",sample.shape)
# print(sample)
print("label shape after to_numpy",label.shape)
# print(label)
print("Merge label")
#group 1 sepsis
#label 1(sepsis) and label 2(SIRS)
for i in range(len(label)):
  if label[i]==2:
    label[i]=1;

print("shape of label",label.shape)
# print(label)
```
